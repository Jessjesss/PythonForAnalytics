```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

1.2 Изучите стоимости недвижимости 

```python
plt.figure(figsize=(20, 8))
sns.histplot(df['price'])
plt.ticklabel_format(style='plain')
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Кол-во');
```

1.3 Изучите распределение квадратуры жилой

```python
plt.hist(df['sqft_living'])
plt.title('Распределение жилой площади')
plt.xlabel('распределение жилой квадратуры')
plt.ylabel('Количество')
```

1.4 Изучите распределение года постройки

```python
plt.hist(df['yr_built'])
plt.title('Распределение года постройки')
plt.xlabel('год постройки')
plt.ylabel('Количество');
```

2.1 Изучите распределение домов от наличия вида на набережную Постройте график Сделайте выводы

```python
df_w = df['waterfront'].value_counts()
df_w
```

```python
plt.hist(df['waterfront'])
plt.title('Распределение домов от наличия вида на набережную')
plt.xlabel('вид на набережную')
plt.ylabel('Количество')
```

Количество домов без вида на набережную больше на 21287.

2.2 Изучите распределение этажей домов

```python
df_fl = df['floors'].value_counts()
df_fl
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x = df_fl.index, y = df_fl)
plt.title('Распределение этажей домов')
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во');
```

Больше всего одноэтажных и двухэтажных домов

2.2 Изучите распределение состояния домов

```python
df_con = df['condition'].value_counts()
df_con
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x = df_con.index, y = df_con)
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Кол-во');
```

Большинство домов со средним состоянием

# 3 задача:

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg') 
```

Чем больше жилплощадь дома, тем выше его стоимость



```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Чем выше оценка дома, тем выше его стоимость



```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Distribution of price by grade');
```

Наибольшую стоимость имеют дома с оценкой конструкции и дизайна 13



```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Distribution of price by floors');
```

Наибольшую стоимость имеют дома с количеством этажей 2,5



```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Distribution of price by view');
```

Наибольшую стоимость имеют дома с оценкой вида 4
